```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\DEEP\Downloads\data.csv")

df.head()
```

```python
df.columns
df.info()
```

```python
df['winner'].value_counts().head(8).plot(kind='bar')
plt.title("Top 8 Most Successful IPL Teams")
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.show()
```

```python
toss_win = df[df['toss_winner'] == df['winner']]
impact = len(toss_win) / len(df)

print("Toss win impact:", round(impact*100, 2), "%")
```

```python
win_type = ['Runs', 'Wickets']
wins = [df['win_by_runs'].astype(bool).sum(),
        df['win_by_wickets'].astype(bool).sum()]

plt.bar(win_type, wins)
plt.title("Win Type Distribution")
plt.show()
```

```python
df['player_of_match'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Players of the Match")
plt.xticks(rotation=45)
plt.show()
```

```python
df['city'].value_counts().head(10).plot(kind='bar')
plt.title("Top Cities Hosting Matches")
plt.xticks(rotation=45)
plt.show()
```
